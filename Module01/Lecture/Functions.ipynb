# Functions
- A block of code that runs only when its called
- Can be passed input (e.g. parameters) and return output

## Defining a function in python:
- Function begins with keyword **def**
- Need parentheses after function name followed by a *:*
- A function can take arguments/parameters
- A function is exited by a **return** statement. Can be used to return value/expression or **None**

```python
def my_function(): #function definition
    #insert code for function to execute
    print('Do stuff')
    return
```

```python
#give the fucntion an argument
def my_function(x):
    print(2*x)
    return
    
my_function(2)
```

# Call by Reference vs. Value

### Pass by Reference
- The address of the vaiable is passed to the function
- Change in the variable also changes its vlalue outside of the function
- Original value is modified



```python
my_list = ['A','B','C']

def func(my_list):
    my_list.append(['D','E'])
    print('my_list inside function:',my_list)
    return

func(my_list)
print('my_list outside the function:',my_list)
```

### Pass by Value
- A copy of the variable is passed to the function
- Does not modify the variable value outside of the function
- Original value is **not** modified



```python
my_list = ['A','B','C']

def func(my_list):
    print('my_list inside function (1):',my_list)
    my_list =(['D','E'])
    print('my_list inside function (2):',my_list)
    return

func(my_list)
print('my_list outside the function:',my_list)
```

## Function Arguments

You can call functions with the following arguments:
- Requiered
- Keyword
- Default
- Variable length

### Required
- Arguments passed to the function in the required order

```python
def f1(x,y):
    print('x = ',x)
    print('y = ',y)
    return

my_var = f1(3,2)
```

### Keyword
- Keywords can be used. This allows arguments to be passed in anyorder.

```python
def f2(name, age):
    print("Name: ", name)
    print("Age: ", age)
    return

f2(age = 25, name = 'Nate')
```

### Default
- Provides a default value if one is not given

```python
def f3(name,age = 35):
    print('Name: ', name)
    print('Age: ', age)
    return
f3(name = 'Nate')
f3(name = 'Nate', age = 40)
```

### Variable Length
- You can use $*$ to pass in an array of parameters as the argument

```python
def my_function(*pars):
    for i in range (0,len(pars)):
        print(pars[i])

my_pars = np.array([6,2,3])
my_function(*my_pars)
```

### We can return a value

```python
def my_function(*pars):
    print("list pars")
    for i in range (0,len(pars)):
        print(pars[i])
    return pars[0]+2*pars[2]

my_pars = np.array([6,2,3])

my_result = my_function(*my_pars)
print("my_result: ", my_result)
```

# Scope
- Depending on where you define a variable it may or may not be accessible at all locations in the program.
- There are two main variable scopes: **global** and **local** variables
- Variables defined within a function body have a **local** scope, and are not accessable outside the function
- Variables defined outside of functions have a **global** scope.

```python
def my_f(y):
    square = y**2
    print('Inside the function square = ', square)
    return square

square = 0;
y = 2;

my_f(y)
print('Outside the function square = ',square)

square = my_f(y)
print('After assignment, outside the function square = ',square)

```
